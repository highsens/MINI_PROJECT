3. product 별로 MA 값을 구한다.
4. product 별로 표준편차를 구한다.

```python
import pandas as pd

kopo_season = pd.read_csv("../dataset/kopo_channel_seasonality_new.csv")
kopo_season.head()
```

```python
sorted_data = kopo_season.sort_values(['REGIONID','PRODUCT','YEARWEEK'])
sorted_data.reset_index(inplace=False, drop=True)
sorted_data.head()
```

```python
group_data = sorted_data.groupby(["REGIONID","PRODUCT","YEARWEEK"])
```

```python
eachdata = group_data.get_group(list(group_data.groups)[0])
```

```python
eachdata
```

### 함수 생성하기

```python
def ma_rolling(eachdata):
    basicData = eachdata.reset_index(drop=True)
#    basicData.head()

    basicData["ma"] = basicData["QTY"].rolling(window=17, center = True, min_periods=1).mean()
    return basicData
```

```python
GroupResult = sorted_data.groupby(["REGIONID","PRODUCT"]).apply(ma_rolling)
```

```python
GroupResult.head()
```

```python
aa = GroupResult.reset_index(drop=True)
```

```python
aa["radio"] = aa["QTY"] / aa["ma"]
aa.head()
```

```python
aa["week"] = aa["YEARWEEK"].astype(str).str[4:]
```
